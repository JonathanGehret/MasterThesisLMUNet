```python
import os

def get_dir_size(path='.'):
    total = 0
    with os.scandir(path) as it:
        for entry in it:
            if entry.is_file():
                total += entry.stat().st_size
            elif entry.is_dir():
                total += get_dir_size(entry.path)
    return total

#print(get_dir_size('data/100k_50_channel_test/'))
print(get_dir_size('data/'))
# 64: 13 120 000 000
# 32: 6 566 400 000
# 16: 3 289 600 000

# 5xmetrics 32 stacked: 32 780 800 000
# 50x those same five metrics stacked: 327 692 800 000 ( I think I lost a few, whoops)
```

```python
# Calculate sum of all values in numpy array
import os
import numpy as np

from torchvision.transforms import ToTensor, Normalize, Compose
from torchvision import transforms

metr_mean = 636003268.469
std = 332168417.904

#             'lsm_l_dcad'   : [636003268.469, 332168417.904]
Normalize = Normalize(mean=metr_mean, std=std)

original_dir = 'data/100k_dcad_5_class_metr/'
#metr_sum = 0

#dir_32 = 'data/100k_dcad_5_class_metr_f32/'

#os.mkdir(dir_32)

# https://discuss.pytorch.org/t/simple-way-to-inverse-transform-normalization/4821/3
reverse_norm = transforms.Compose([transforms.Normalize(mean = 0, std = 1/std),
                    transforms.Normalize(mean = -1 * metr_mean, std = 1),])


for metr_name in os.listdir(original_dir):
    
    # load metric
    metr = np.load(original_dir + metr_name)
    
    # convert to tensor for normalizatoin and then yck to np
    metr_tensor = ToTensor()(metr)
    metr_norm = Normalize(metr_tensor)
    metr_norm_np = metr_norm.numpy()
    
    # float 64 
    print('64', metr_name, np.amax(metr))
    print('64', metr_name, 'norm', np.amax(metr_norm_np))
    
    # reversed 64
    rev_norm_64 = reverse_norm(metr_norm)
    rev_norm_np_64 = rev_norm_64.numpy()
    print('64', metr_name, 'rev', np.amax(rev_norm_np_64))
    
    
    # float 32
    metr = np.float32(metr)
    print('32', metr_name, np.amax(metr))
    metr_norm_np = np.float32(metr_norm_np)
    print('32', metr_name, 'norm', np.amax(metr_norm_np))
    
    # reverse norm 32
    metr_tensor_32 = ToTensor()(metr_norm_np)
    rev_norm_32 = reverse_norm(metr_tensor_32)
    rev_norm_np_32 = rev_norm_32.numpy()
    print('32', metr_name, 'rev', np.amax(rev_norm_np_32))
    
    # float 16
    metr = np.float16(metr)
    print('16', metr_name, np.amax(metr))
    metr_norm_np = np.float16(metr_norm_np)
    print('16', metr_name, 'norm', np.amax(metr_norm_np))
    
    # reverse norm 16
    norm_16_to_32 = np.float32(metr_norm_np)
    metr_tensor_16 = ToTensor()(norm_16_to_32)
    rev_norm_16 = reverse_norm(metr_tensor_16)
    rev_norm_np_16 = rev_norm_16.numpy()
    print('16', metr_name, 'rev', np.amax(rev_norm_np_16))
    
    
    
```
